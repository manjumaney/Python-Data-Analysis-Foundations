Data Analysis with Python



### Python Language Basics

#### 1. Language Semantics

```python
a = [10, 20, 30]
b = a

print(a)
print(b)

print(id(a))
print(id(b))

b.append(40)

print("After modifying b:")
print(a)
print(b)

b = [11, 12]

print("Now a = ", a)
print("b =",b)

print(id(a))
print(id(b))

```

```python
#Lists vs Tuples

my_list = [5, 6, 7]
my_tuple = (5, 6, 7)

my_list[0] = 100
print(my_list)

# This will cause an error (expected)
#my_tuple[0] = 100
```

```python
#copying objects
original = [1, 2, [3, 4]]
copy_list = original.copy()

copy_list[0] = 99
copy_list[2].append(5)

print("Original:", original)
print("Copy:", copy_list)
```

```python
#Functions

def add_value(lst):
    lst.append(50)

numbers = [10, 20, 30]
add_value(numbers)

print(numbers)
```

```python
#operators == vs is
x = [1, 2, 3]
y = [1, 2, 3]
z = x

print(x == y)   # value comparison
print(x is y)   # identity comparison

print(x is z)
```

#### 2. Scalar Types

```python
#Integers and Floats

a = 10
b = 3.5

print(type(a))
print(type(b))

c = a + b
print(c, type(c))
```

```python
#Type casting

x = "25"
y = int(x)

z = float(y)

print(y, type(y))
print(z, type(z))
```

```python
#Boolean Type
is_active = True
is_complete = False

print(type(is_active))
print(is_active and is_complete)
print(is_active or is_complete)
```

```python
#None Type

value = None

print(value)
print(type(value))
```

```python
#Comparisons
m = 10
n = 20

print(m < n)
print(m == n)
print(m != n)
```

```python
#Numeric operators

p = 7
q = 2

print(p / q)    # division
print(p // q)   # floor division
print(p ** q)   # power
```

#### 3. Control Flow

```python
#if / elif / else

score = 78

if score >= 80:
    result = "Distinction"
elif score >= 60:
    result = "Pass"
else:
    result = "Fail"

print(result)
```

```python
#FOR Loop

numbers = [2, 4, 6, 8]

for num in numbers:
    print(num * 2)
```

```python
#WHILE Loop

count = 1

while count <= 5:
    print(count)
    count += 1
```

```python
# BREAK and CONTINUE

for i in range(1, 6):
    if i == 3:
        continue
    if i == 5:
        break
    print(i)
```

```python
#RANGE()

for i in range(0, 10, 2):
    print(i)
```

```python
#Conditional expressions

age = 20
status = "Adult" if age >= 18 else "Minor"
print(status)
```

```python
#Exception Handling

try:
    value = int("abc")
except ValueError:
    print("Conversion failed")
```
