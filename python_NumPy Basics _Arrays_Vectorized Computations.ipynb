#### 4. Array-Oriented Programming

```python
#Vectorized Computation
import numpy as np

values = np.array([1, 2, 3, 4, 5])

result = values * 10
print(result)

```

```python
#Comparing Loop vs Vectorized

# Using loop
loop_result = []
for v in values:
    loop_result.append(v * 10)

print(loop_result)
```

```python
# Using NumPy vectorization
vectorized_result = values * 10
print(vectorized_result)
```

```python
#Conditional Logic with Arrays

scores = np.array([45, 72, 88, 60])

status = np.where(scores >= 60, "Pass", "Fail")
print(status)
```

```python
#Mathematical Expressions

x = np.array([1, 2, 3])
y = np.array([4, 5, 6])

expr = (x * 2) + (y / 2)
print(expr)
```

#### 6. Linear Algebra

```python
#Matrix creation


A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])
```

```python
# Matrix Addition & Multiplication
print(A + B)
print(A @ B)
```

```python
#Transpose
print(A.T)
```

```python
#Determinant

det_A = np.linalg.det(A)
print(det_A)
```

```python
#Inverse

inv_A = np.linalg.inv(A)
print(inv_A)
```

```python
#Solving Linear Equations

b = np.array([5, 11])

solution = np.linalg.solve(A, b)
print(solution)
```

#### 7. Random Walk

```python
#Simple Random Walk (1D)

np.random.seed(0)

steps = np.random.choice([1, -1], size=20)
walk = steps.cumsum()

print("Steps:", steps)
print("Random Walk:", walk)
```

```python
#Random Walk with Conditions

steps = np.random.choice([1, -1], size=50)
walk = steps.cumsum()

max_value = walk.max()
min_value = walk.min()

print("Max value:", max_value)
print("Min value:", min_value)
```

```python
# Find First Crossing Point
crossing_point = np.where(np.abs(walk) >= 5)[0]

if len(crossing_point) > 0:
    print("First crossing index:", crossing_point[0])
else:
    print("Threshold not crossed")
```

```python
#Multiple Random Walks
nwalks = 5
nsteps = 30

walks = np.random.choice([1, -1], size=(nwalks, nsteps))
walks = walks.cumsum(axis=1)

print(walks)
```

```python
#Count Walks Crossing a Threshold

hits = np.abs(walks) >= 10
num_hits = hits.any(axis=1)

print("Walks crossing threshold:", num_hits.sum())
```
