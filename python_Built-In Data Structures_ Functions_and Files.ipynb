## Data Analysis with Python

### Built-In Data Structures, Functions, and Files

#### 1. Built-In Data Structure, Sequences

```python
#Lists

numbers = [10, 20, 30, 40]

numbers.append(50)
numbers.remove(20)

print(numbers)
```

```python
numbers.sort(reverse=True)
print(numbers)
```

```python
# Tuples

point = (3, 4)
print(point[0])
```

```python
x, y = point
print(x, y)
```

```python
#Dictionaries

student = {
    "name": "Alex",
    "age": 22,
    "program": "Data Analytics"
}

print(student["name"])
```

```python
student["age"] = 23
student["city"] = "Toronto"

print(student)
```

```python
#Sets

values = {1, 2, 3, 3, 4}
print(values)
```

```python
values.add(5)
values.remove(2)
print(values)
```

```python
#List Comprehensions

squares = [x**2 for x in range(1, 6)]
print(squares)
```

```python
even_numbers = [x for x in range(10) if x % 2 == 0]
print(even_numbers)
```

#### 2. Functions

```python
#Defining a function

def greet(name):
    return f"Hello, {name}"

print(greet("Sam"))
```

```python
def add_numbers(a, b):   # multiple parameter
    return a + b

print(add_numbers(5, 7))
```

```python
def power(base, exponent=2):   #default arguments
    return base ** exponent

print(power(4))
print(power(4, 3))
```

```python
def student_info(name, age, program): 
    return f"{name}, {age}, {program}"

print(student_info(age=21, program="AI", name="Riya"))  #keyword arguments
```

```python
def calculate(a, b):
    return a + b, a * b

sum_val, product_val = calculate(3, 4)
print(sum_val, product_val) #returning multiple values
```

```python
#Lamda functions

multiply = lambda x, y: x * y
print(multiply(6, 7))
```

```python
#Functions as arguemnts

def apply_operation(func, x, y):
    return func(x, y)

print(apply_operation(add_numbers, 10, 5))
```

```python
#Docstrings

def subtract(a, b):
    """
    Returns the difference between two numbers
    """
    return a - b

print(subtract(10, 3))
print(subtract.__doc__)
```

#### 3. Files & Operating System

```python
#writing to a file
file = open("sample.txt", "w")
file.write("This is a sample file.\nPython file handling.")
file.close()
```

```python
#reading from a file
file = open("sample.txt", "r")
content = file.read()
file.close()

print(content)
```

```python
#reading line by line

file = open("sample.txt", "r")

for line in file:
    print(line.strip())

file.close()
```

```python
#working with OS module

import os

print(os.getcwd())
```

```python
#creating and removing directories

os.mkdir("test_folder")
print(os.listdir())

os.rmdir("test_folder")
print(os.listdir())
```

```python
#checking file existence

if os.path.exists("sample.txt"):
    print("File exists")
else:
    print("File not found")
```

```python
#File path

path = os.path.join(os.getcwd(), "sample.txt")
print(path)
```
